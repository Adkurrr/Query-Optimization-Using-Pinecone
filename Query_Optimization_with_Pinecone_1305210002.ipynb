#Melakukan Query

```python
queries = [
    "survival game series",
    "Lee Jung-jae series",
    "best TV series of 2021",
    "dark dystopian series with intense plot",
    "emotional rollercoaster series",
    "better than Squid Game",
    "series with unpredictable plot twists like Squid Game"
]
```

```python
query_results_no_cache = {}
total_start_time_no_cache = time.time()

for query in queries:
    start_time = time.time()
    query_embedding = model.encode([query]).astype('float32')
    results = index.query(vector=query_embedding.tolist(), top_k=3)
    end_time = time.time()

    query_results_no_cache[query] = {
        'results': results['matches'],
        'execution_time': end_time - start_time
    }

total_end_time_no_cache = time.time()
total_execution_time_no_cache = total_end_time_no_cache - total_start_time_no_cache
```

```python
print("Results without caching:\n")
for query, result in query_results_no_cache.items():
    print(f"Query: {query}")
    print("Results:")
    for match in result['results']:
        review_id = match['id']
        score = match['score']
        review_body = doc.iloc[int(review_id.split('-')[1])]['Review_body']
        print(f"ID: {review_id}, Score: {score}, Review: {review_body}")
    print(f"Execution time: {result['execution_time']} seconds")
    print("\n")

print(f"Total execution time without caching: {total_execution_time_no_cache} seconds\n")
```

#Caching

```python
cache = {}

def query_with_cache(query):
    if query in cache:
        return cache[query], 0  # 0 execution time for cache hit
    start_time = time.time()
    query_embedding = model.encode([query]).astype('float32')
    results = index.query(vector=query_embedding.tolist(), top_k=3)
    end_time = time.time()
    cache[query] = results
    return results, end_time - start_time

query_results_cache = {}
total_start_time_cache = time.time()

for query in queries:
    results, execution_time = query_with_cache(query)
    query_results_cache[query] = {
        'results': results['matches'],
        'execution_time': execution_time
    }

total_end_time_cache = time.time()
total_execution_time_cache = total_end_time_cache - total_start_time_cache
```

```python
print("Results with caching:\n")
for query, result in query_results_cache.items():
    print(f"Query: {query}")
    print("Results:")
    for match in result['results']:
        review_id = match['id']
        score = match['score']
        review_body = doc.iloc[int(review_id.split('-')[1])]['Review_body']
        print(f"ID: {review_id}, Score: {score}, Review: {review_body}")
    print(f"Execution time: {result['execution_time']} seconds")
    print("\n")

print(f"Total execution time with caching: {total_execution_time_cache} seconds")
```
